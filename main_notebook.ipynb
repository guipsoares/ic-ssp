```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
from ipyleaflet import Map, Marker, Popup, FullScreenControl
from ipywidgets import HTML
```

```python

meses = []

for mes in os.listdir('./dados_brutos_2018/furto_celular/'):
    meses.append(mes)
    
todos_meses = pd.DataFrame()

for mes in meses:
    df = pd.read_excel('./dados_brutos_2018/furto_celular/' + mes)
    todos_meses = pd.concat([todos_meses, df])
    
todos_meses.to_csv("dados_ano_todo_2018.csv", index= False)
```

```python

df = pd.read_csv('./tabela.csv')

df = pd.DataFrame(df)

```

```python



df = df.dropna(subset=['LATITUDE', 'LONGITUDE'])


df = df.drop(df.columns[0], axis = 1)
pd.set_option('display.max_columns', None)
results = df.groupby('MES_BO').sum()


df
```

```python
center = (-22.0027, -47.8986) ##USP sao carlos

m = Map(center = center, zoom = 13)
m.add_control(FullScreenControl())

lat = df['LATITUDE'].values
long = df['LONGITUDE'].values
info1 = df['PERIDOOCORRENCIA'].values

i = 0

for i in range(0, len(lat)):
    marker = Marker(location = (lat[i], long[i]), draggable = False)
    m.add_layer(marker)
    popup = Popup(max_width = 500, min_width = 10)
    mensagem1 = HTML()
    mensagem1.value = 'Periodo: ' + info1[i]
    marker.popup = mensagem1

m
```
